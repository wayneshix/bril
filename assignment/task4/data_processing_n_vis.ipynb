```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('/Users/way/projects/bril/assignment/task3/out.csv')




desired_order = ['benchmark', 'baseline', 'tdce', 'global_dce', 'global_constant_folding_and_dce', 'loop_optimization'] 

# Pivot the DataFrame
pivot_table = df.pivot(index='benchmark', columns='run', values='result')

# Reset index to make 'benchmark' a column
pivot_table.reset_index(inplace=True)

# Now apply the desired order
pivot_table = pivot_table[desired_order]

# Replace NaN with empty strings if needed
pivot_table = pivot_table.fillna('')
```

```python
pivot_table = pivot_table[~pivot_table.apply(lambda row: row.astype(str).str.contains('incorrect|timeout').any(), axis=1)]
```
